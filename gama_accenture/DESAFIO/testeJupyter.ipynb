```python
query = '''
SELECT  *
FROM fato_resultado fato
INNER JOIN dim_piloto piloto
ON fato.id_piloto = piloto.id_piloto
'''

df_teste = pd.read_sql_query(query,engine)
df_teste.head()
```

```python
# Criando um DataFrame a partir de uma query utilizando parâmetros dinâmicos.

query = 'SELECT *' \
      'FROM dim_construtor ' \
      'WHERE nome_construtor = %s'
df3 = pd.read_sql_query(query, engine,params=[input('digite um construtor: \n')])
df3.head()
```
